##### Merge the three dataframes

```python
transitional_df = df.merge(codes_df, how='left', left_on='Country', right_on='Country or Area')
Merged_df = transitional_df.merge(continents_df, how='left', left_on='ISO-alpha3 code', right_on='Three_Letter_Country_Code')
```

##### Drop New Useless Columns

```python
Merged_df = Merged_df.drop(['Country or Area', 'Region', 'Three_Letter_Country_Code',] , axis=1)
```

##### Rearranging columns so that the dataframe is read easily

```python
ordered_columns = ['Country', 'Continent_Name', 'ISO-alpha3 code', 'Population', 'Net migration',
                   'Infant mortality (per 1000 births)', 'GDP ($ per capita)',
                   'Literacy (%)', 'Birthrate', 'Deathrate',
                   'Agriculture', 'Industry', 'Service']
```

```python
Merged_df = Merged_df.reindex(columns=ordered_columns)
```

```python
@app.callback(
    
    # Plots
    Output(component_id='map_graph', component_property='figure'),
    Output(component_id='scatter_plot', component_property='figure'),
    Output(component_id='bar_graph', component_property='figure'),
    Output(component_id='pie_chart', component_property='figure'),
    
    # Bans Info
    Output(component_id='Birthrate_Value', component_property='children'),
    Output(component_id='Birthrate_Area', component_property='children'),
    Output(component_id='Deathrate_Value', component_property='children'),
    Output(component_id='Deathrate_Area', component_property='children'),
    Output(component_id='Infant_Mortality_Value', component_property='children'),
    Output(component_id='Infant_Mortality_Area', component_property='children'),
    
    # Plots Headers
    Output(component_id='map_graph_title', component_property='children'),
    Output(component_id='scatter_plot_title', component_property='children'),
    Output(component_id='bar_graph_title', component_property='children'),
    Output(component_id='pie_chart_title', component_property='children'),
    
    
    # Inputs
    Input(component_id='continent_dropdown', component_property='value'),
    Input(component_id='map_graph', component_property='hoverData'),
    Input(component_id='scatter_plot', component_property='hoverData'),

)


def my_function(continent_dropdown, map_hover, scatter_hover):
    
    # Bar Plot
    hover_country_bar = 'Egypt'  
    bar_values=[]
    
    if map_hover:
         hover_country_bar = map_hover['points'][0]['customdata'][0]

    bar_df = Merged_df[Merged_df['Country'] == hover_country_bar]

    bar_labels = ['Birthrate','Deathrate','Infant mortality']
    
    for i in bar_labels:
        bar_values.append(float(bar_df[i]))
            
    bar_fig = px.bar(x=bar_labels, y=bar_values, color=bar_labels,
                         color_discrete_map={'Birthrate':'#009d00','Deathrate':'black', 'Infant mortality':'#dc0043'})
    bar_fig.update_layout(showlegend = False)
    bar_fig.update_yaxes(title='')
    bar_fig.update_xaxes(title='')
    
    
    # Pie Chart
    hover_country_pie = 'Egypt'
    pie_values=[]
    
    if scatter_hover:
        hover_country_pie = scatter_hover['points'][0]['customdata'][0]
    
    pie_df = Merged_df[Merged_df['Country'] == hover_country_pie]

    pie_labels = ['Agriculture','Industry','Service']
    
    for i in pie_labels:
        pie_values.append(float(pie_df.loc[:,i]))
    
    pie_fig = px.pie(names=pie_labels, values=pie_values, color=pie_labels,
                         color_discrete_map={'Agriculture':'#57bebb','Industry':'#8864a9', 'Service':'#047d7c'})

    
    # Plots Header
    map_graph_title = '{} Countries Population'.format(continent_dropdown.title())
    scatter_plot_title = 'Literacy vs Net Migration in {}'.format(continent_dropdown.title())
    bar_graph_title = '{} Rates'.format(hover_country_bar)
    pie_chart_title = '3 Main Pillers of {}'.format(hover_country_pie)
    
    
    
    if continent_dropdown == 'world':
        
        # BANS
        Birthrate_Value = str("%.1f" %Merged_df['Birthrate'].mean())
        Deathrate_Value = str("%.1f" %Merged_df['Deathrate'].mean())
        Infant_Mortality_Value = str("%.1f" %Merged_df['Infant mortality'].mean())
        Area = 'a year Wolrdwide'
        
        
        # Map Plot
        map_fig = px.choropleth(Merged_df, locations="ISO-alpha3 code", color="Log_Population",
                  hover_name = 'Country', hover_data = ['Population'], custom_data=['Country'],
                  scope='world', color_continuous_scale='Bluyl')
        
        
        # Scatter Plot
        scatter_fig = px.scatter(data_frame=Merged_df,
                                 x='Literacy (%)', y='Net migration', color='Country',
                                 hover_name = 'Country', custom_data=['Country'])
        scatter_fig.update_layout(showlegend = False)
        
        return map_fig, scatter_fig, bar_fig, pie_fig,\
               Birthrate_Value, Area, Deathrate_Value, Area, Infant_Mortality_Value, Area,\
               map_graph_title, scatter_plot_title, bar_graph_title, pie_chart_title

    
    else:
        
        new_df = Merged_df[Merged_df['Continent_Name'] == continent_dropdown]

        # BANS
        Birthrate_Value = str("%.1f" %new_df['Birthrate'].mean())
        Deathrate_Value = str("%.1f" %new_df['Deathrate'].mean())
        Infant_Mortality_Value = str("%.1f" %new_df['Infant mortality'].mean())
        Area = 'a year ' + continent_dropdown.title()
        
        
        # Map Plot
        map_fig = px.choropleth(data_frame=new_df, locations="ISO-alpha3 code", color="Log_Population",
                            hover_name = 'Country', hover_data = ['Population'], custom_data=['Country'],
                            scope = continent_dropdown, color_continuous_scale='Bluyl')
        
        
        # Scatter Plot
        scatter_fig = px.scatter(data_frame=new_df, x='Literacy (%)', y='Net migration', color='Country',
                                 hover_name = 'Country', custom_data=['Country'])
        scatter_fig.update_layout(showlegend = False)
        
    return map_fig, scatter_fig, bar_fig, pie_fig,\
               Birthrate_Value, Area, Deathrate_Value, Area, Infant_Mortality_Value, Area,\
               map_graph_title, scatter_plot_title, bar_graph_title, pie_chart_title
```
